```python
# lines
#words
#characters

from functools import reduce
with open('file_handling/poem.txt','r') as f:
    lines = f.readlines() #splitlines()
    print(f'{len(lines)} Lines')
    f.seek(0)
    words = list(filter(lambda x:x!='\n',f.read().split(' ')))
    print(f'{len(words)} Words')
    f.seek(0)
    chars = len(f.read())
    print(f'{(chars)} characters')
```

```python
from functools import reduce
f=open('file_handling/poem.txt','r')
print("Count of sentences in the file :",len(f.read().split("."))-1)
print("-"*50)
f.seek(0)
f.read()
print("Number of characters in the file are :",f.tell())
print("-"*50)
f.seek(0)
list_temp_words=f.readlines()
list_words=[]
for i in list_temp_words:
    list_eles=i.split(" ")
    #print(list_eles)
    empty_ch=list_eles.count("")
    newline_ch=list_eles.count("\n")
    #print(list_eles.count(""))
    list_words.append(len(list_eles)-empty_ch-newline_ch)  
#print(list_words) 
print("Number of words in the files ",reduce((lambda a,b : a+b),list_words))
```

# Regex
- RegEx can be used to check if a string contains the specified search pattern.
- Python has a built-in package called re,

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)      # mostle used
\D      - Not a Digit (0-9)# mostle used
\w      - Word Character (a-z, A-Z, 0-9, _)# mostle used
\W      - Not a Word Character             # mostle used 
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary # mostle used
\B      - Not a Word Boundary# mostle used
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More # mostle used
+       - 1 or More # mostle used
?       - 0 or One  # mostle used
{3}     - Exact Number # mostle used
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```

```python
# Word Boundary

string = 'cat catherine catholic copycat concat uncatchable'
# Word Boundary starting
# What you want to search
pattern = re.compile(r'\bcat') 

# From which string you want to extract the data
match = pattern.finditer(string)

for i in match:
    print(i)
```

```python
# Word Boundary

string = 'cat catherine catholic copycat concat uncatchable'

# cat\b matches the string "cat" when it appears at the end of a word. Specifically, it ensures
#that "cat" is either followed by a non-word character or is at the end of the string.
#It does not care about what comes before "cat".

# What you want to search
pattern = re.compile(r'cat\b')

# From which string you want to extract the data
match = pattern.finditer(string)

for i in match:
    print(i)
```

```python
# Word Boundary

string = 'cat catherine catholic copycat concat uncatchable'
# \bcat\b ensures that "cat" is matched only when it is a standalone word.
# What you want to search
pattern = re.compile(r'\bcat\b')

# From which string you want to extract the data
match = pattern.finditer(string)

for i in match:
    print(i)
```

```python
# Word Boundary

string = 'cat catherine catholic copycat concat uncatchable'

# What you want to search
pattern = re.compile(r'\Bcat')

# From which string you want to extract the data
match = pattern.finditer(string)

for i in match:
    print(i)
```

```python
# Word Boundary

string = 'cat catherine catholic copycat concat uncatchable dddcatddd '
#string = 'uncatchable dddcatddd '
# B Not a Word Boundary last 2
# What you want to search
#\Bcat\B matches the string "cat" only when "cat" is not at the start or end of a word. This means that "cat" must be part of a larger word and surrounded 
#by word characters (letters, digits, or underscores) on both sides.

pattern = re.compile(r'\Bcat\B')

# From which string you want to extract the data
match = pattern.finditer(string)

for i in match:
    print(i)
```

```python
# This is a capturing group (r|s) that matches either "r" or "s".
#The s? matches the character "s" zero or one time.
#\s matches any whitespace character (such as space, tab, newline).


# What you want to search
pattern = re.compile('M(r|s)s?\.?[\s_]?[A-Za-z]*')

# From which string you want to extract the data
match = pattern.finditer(text_to_search)

for i in match:
    print(i)
```
