# Reshape 
- Reshaping means changing the shape of an array.

```python
lst = [[1,2], [6,7],[6,7],[6,7]]

arr = np.array(lst)

print(arr)
```

```python
arr.shape
```

```python
arr.size
```

```python
arr.reshape((4,2)) # 4 rows 2 columns
```

```python
np.arange(24).reshape((3,8)) 
```

```python
np.arange(24).reshape((2,6,2)) #    d,r,c one matrix behind other
```

# infer

```python
# rows,  c-`-1` indicates that NumPy should automatically determine the number of columns 
#so that the total number of elements remains the same 
np.arange(24).reshape((4,-1)) # rows 24/4 = 6 columns
```

```python
# reshapes this array into a 3-dimensional array with 4 blocks, each containing 2 rows. The `-1` means NumPy
#automatically calculates the size of the last dimension.
np.arange(24).reshape((4,2,-1)) # depth-4blocks 2 rows 4*2*3
```

```python
np.arange(24).reshape((4,-1,2)) # 4matrix, rows ? columns 2 , 24/2 12 rows
```

```python
np.arange(24).reshape((-1,4,-1))
```

```python
np.arange(36).reshape(6,6) # 6 rows, 6 columns
```
