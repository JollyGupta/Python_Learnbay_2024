# Slicing can be done by 3 methods

1. Normal method

2. loc

3. iloc 

```python
new_df = df.head(10)
e_list = []

for i in new_df.index:
    e_list.append(f'row_{i}')

new_df.index = e_list

new_df
```

# Normal slicing

- slicing on columns is not allowed

1. zero indexing
- start is inclusive
- stop is exclusive

2. custom indexing
- start is inclusive
- stop is inclusive

```python
new_df[2 : 5] # store it into df then print 
```

```python
new_df[5 : 8]
```

```python
new_df['row_2' : 'row_5']
```

# loc method

- slicing on rows and columns are allowed

1. custom indexing
- start is inclusive
- stop is inclusive

```python
new_df.loc[2 : 5]
```

```python
new_df.loc['row_2' : 'row_5']
```

```python
new_df.loc['row_2' : 'row_5' , 'start station id' : 'end station name']
```

```python
new_df.loc['row_1' : 'row_3' :2 , 'starttime' : 'end station id' : 2]
```

```python
df[2:4]
```

```python
df.loc[2:4]
```

```python
df.loc[10 : 30 : 10 , ::2]
```

```python
df
```

# iloc method

- slicing on rows and columns are allowed

1. zero indexing
- start is inclusive
- stop is exclusive

```python
new_df.iloc['row_2' , 'row_4']
```

```python
new_df.iloc[2:5]
```

```python
new_df.iloc[2:5 , 3:7]
```
